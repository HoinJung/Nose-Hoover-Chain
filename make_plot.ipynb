```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation, rc
import os 
import math
%matplotlib inline

"""
Call result of simulation each Q.
These files have the data of Potential Energy, Kinetic Energy, Total Energy, Temperature, Preesure for each step.
"""

data1 = pd.read_csv('./result/result_q0.1.csv')
data2 = pd.read_csv('./result/result_q0.5.csv')
data3 = pd.read_csv('./result/result_q1.0.csv')
data4 = pd.read_csv('./result/result_q2.0.csv')
data5 = pd.read_csv('./result/result_q5.0.csv')

# convert dataframe to array
data1_arr = data1.values
data2_arr = data2.values
data3_arr = data3.values
data4_arr = data4.values
data5_arr = data5.values
```

```python
"""
Make graph of Kinetic Energy.
The y range is calculated by minimum, maximum of the result of Q=0.1
"""

# set figures X range
N=len(data1_arr[:,1])
X_min, X_max = data1_arr[0,1], data1_arr[N-1,1]
X = np.zeros(N)
X[:] = np.linspace(X_min,X_max,N)
stepx=1000

# set figures Y range (Kinetic Energy)
Y_min= math.floor(min(data1_arr[:,3])) - math.floor(min(data1_arr[:,3]))%10
Y_max= math.ceil( max(data1_arr[:,3])) - math.ceil( max(data1_arr[:,3]))%10
stepy= (Y_max - Y_min)/10

# setup plot to figure
plt.rcParams['figure.figsize'] = (12,8)
plt.rcParams.update({'font.size': 15})

# title of plot
plt.title('Kinetic Energy')

# plotting result graph
plt.xlabel('steps')  
plt.ylabel('Energy') 
plt.xlim(X_min,X_max)
plt.ylim(Y_min,Y_max)
plt.xticks(np.arange(X_min, X_max,step=stepx))
plt.yticks(np.arange(Y_min,Y_max,step=stepy))
plt.tick_params(axis='x', direction='inout', length=10, pad=6, labelsize=14)
plt.tick_params(axis='y', direction='inout', length=10, pad=15, labelsize=12,)
plt.plot(X[:], data1_arr[:,3],'-b',label='Q=0.1')
plt.plot(X[:], data2_arr[:,3],'-r',label='Q=0.5')
plt.plot(X[:], data3_arr[:,3],'-g',label='Q=1.0')
plt.plot(X[:], data4_arr[:,3],'-m',label='Q=2.0')
plt.plot(X[:], data5_arr[:,3],'-y',label='Q=5.0')
plt.grid(True)
plt.legend()
fig1=plt.gcf()
plt.show()
plt.draw()
if not os.path.exists('result_graph'):
    os.makedirs('result_graph')   
fig1.savefig('result_graph/'+'result_fig_'+'KE'+'.png',dpi=100)
```

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation, rc
import os 
import math

%matplotlib inline

"""
Call velocity data for each Q and each paritlcles.
I will make a plot instant velocity
"""

vel_1 = pd.read_csv('./result/vel_result_q0.1.csv')
vel_2 = pd.read_csv('./result/vel_result_q0.5.csv')
vel_3 = pd.read_csv('./result/vel_result_q1.0.csv')
vel_4 = pd.read_csv('./result/vel_result_q2.0.csv')
vel_5 = pd.read_csv('./result/vel_result_q5.0.csv')

# convert dataframe to array
vel1_arr = vel_1.values
vel2_arr = vel_2.values
vel3_arr = vel_3.values
vel4_arr = vel_4.values
vel5_arr = vel_5.values
```

```python
"""
Initialize parameters
And plot velocity ditribution for z-axis
"""

# setup instant velocity 
nx=2998
ny=2999
nz=3000

# number of gird
n=1000

# number of histogram
hist = 12

# initialize X range and Y range 
X = np.linspace(-3,3,n)
Y = np.zeros(n)

# setup plot to figure
plt.rcParams['figure.figsize'] = (12,8)
plt.rcParams.update({'font.size': 15})

# title of plot
plt.title('Velocity Distribution for Z-axis')

# plotting result graph
plt.xlabel('Velocity')  
plt.ylabel('Probability') 

# initialize range of histogram
bins = np.linspace(-3,3,hist)

# Calculate Maxwell-Boltzmann distribution
Y = (1/math.sqrt(2*math.pi)) * (np.exp((-X**2)/2))

# make a array of instant z_velocity for all Q
v1z =vel1_arr[:,nz].astype(np.float)
v2z =vel2_arr[:,nz].astype(np.float)
v3z =vel3_arr[:,nz].astype(np.float)
v4z =vel4_arr[:,nz].astype(np.float)
v5z =vel5_arr[:,nz].astype(np.float)

# make a histogram array for all Q
histogram1z = np.histogram(v1z, bins=bins, density=True)[0]
histogram2z = np.histogram(v2z, bins=bins, density=True)[0]
histogram3z = np.histogram(v3z, bins=bins, density=True)[0]
histogram4z = np.histogram(v4z, bins=bins, density=True)[0]
histogram5z = np.histogram(v5z, bins=bins, density=True)[0]


# re-arrange X axis of histogram to plot right
bins = 0.5*(bins[1:] + bins[:-1])

# plot exact(Maxwell-Boltzmann distribution) and velocity distribution for all Q
import seaborn as sns
plt.plot(X,Y,'--k',label='Maxwell-Boltzmann')
plt.plot(bins, histogram1z,'-b',label='Q=0.1') 
plt.plot(bins, histogram2z,'-r',label='Q=0.5') 
plt.plot(bins, histogram3z,'-g',label='Q=1.0')
plt.plot(bins, histogram4z,'-m',label='Q=2.0') 
plt.plot(bins, histogram5z,'-y',label='Q=5.0')
plt.legend()
fig1=plt.gcf()
plt.show()
plt.draw()
if not os.path.exists('result_graph'):
    os.makedirs('result_graph')   
fig1.savefig('result_graph/'+'result_fig_'+'Velocity_Z'+'.png',dpi=100)
```

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os 
import math

%matplotlib inline

"""
Call coordinate data for all Q and paritlcles.
I will make a scatter plot of instant coordinate
"""

coord_1 = pd.read_csv('./result/coord_result_q0.1.csv')
coord_2 = pd.read_csv('./result/coord_result_q0.5.csv')
coord_3 = pd.read_csv('./result/coord_result_q1.0.csv')
coord_4 = pd.read_csv('./result/coord_result_q2.0.csv')
coord_5 = pd.read_csv('./result/coord_result_q5.0.csv')

# convert dataframe to array
coord1_arr = coord_1.values
coord2_arr = coord_2.values
coord3_arr = coord_3.values
coord4_arr = coord_4.values
coord5_arr = coord_5.values
```

```python
# select the step(moment) to check coordinate
step_number = 30
x_count = step_number*3 - 2
y_count = step_number*3 - 1
z_count = step_number*3
Q = 1.0
# make a array of instant coordinate for selected Q
if Q==0.1 : 
    coord_x =coord1_arr[:,x_count].astype(np.float)
    coord_y =coord1_arr[:,y_count].astype(np.float)
    coord_z =coord1_arr[:,z_count].astype(np.float)
elif Q==0.5 : 
    coord_x =coord2_arr[:,x_count].astype(np.float)
    coord_y =coord2_arr[:,y_count].astype(np.float)
    coord_z =coord2_arr[:,z_count].astype(np.float)
elif Q==1.0 : 
    coord_x =coord3_arr[:,x_count].astype(np.float)
    coord_y =coord3_arr[:,y_count].astype(np.float)
    coord_z =coord3_arr[:,z_count].astype(np.float)
elif Q==2.0 : 
    coord_x =coord4_arr[:,x_count].astype(np.float)
    coord_y =coord4_arr[:,y_count].astype(np.float)
    coord_z =coord4_arr[:,z_count].astype(np.float)
elif Q==5.0 : 
    coord_x =coord5_arr[:,x_count].astype(np.float)
    coord_y =coord5_arr[:,y_count].astype(np.float)
    coord_z =coord5_arr[:,z_count].astype(np.float)
else :
    print("wrong Q selected ")


plt.rcParams['figure.figsize'] = (12,8)
plt.rcParams.update({'font.size': 15})

fig = plt.figure()

ax=fig.gca(projection='3d')
ax.scatter(coord_x,coord_y,coord_z, c='b')
ax.set_xlabel("X coordinate")
ax.set_ylabel("Y coordinate")
ax.set_zlabel("Z coordinate")
plt.title('3D Coordinates of Step No.%i (Q=%.1f)' %(step_number*10,Q))
fig1=plt.gcf()
plt.show()
plt.draw()
if not os.path.exists('result_graph'):
    os.makedirs('result_graph')   
fig1.savefig('result_graph/'+'result_fig_'+'coordinates_step_'+str(step_number)+'_'+'.png',dpi=100)
```

```python

def scatter_hist(x, y, ax, ax_histx, ax_histy):
    # no labels
    ax_histx.tick_params(axis="x", labelbottom=False)
    ax_histy.tick_params(axis="y", labelleft=False)

    # the scatter plot:
    ax.scatter(x, y, c='b')

    # now determine nice limits by hand:
    binwidth = 0.25
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth) + 1) * binwidth

    bins = np.arange(0, lim + binwidth, binwidth)
    ax_histx.hist(x, bins=bins,color='dodgerblue')
    ax_histy.hist(y, bins=bins,color='dodgerblue', orientation='horizontal')

# definitions for the axes
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
spacing = 0.005


rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom + height + spacing, width, 0.2]
rect_histy = [left + width + spacing, bottom, 0.2, height]

# start with a square Figure
fig = plt.figure(figsize=(8, 8))

ax = fig.add_axes(rect_scatter)
ax_histx = fig.add_axes(rect_histx, sharex=ax)
ax_histy = fig.add_axes(rect_histy, sharey=ax)

# use the previously defined function
scatter_hist(coord_x, coord_y, ax, ax_histx, ax_histy)
fig1=plt.gcf()
plt.show()
plt.draw()
if not os.path.exists('result_graph'):
    os.makedirs('result_graph')   
fig1.savefig('result_graph/'+'result_fig_'+'coordinates_dist_step_'+str(step_number)+'_'+'.png',dpi=100)
```
